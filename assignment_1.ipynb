### Find the car (or cars) with the best ratio of horsepower to weight among all cars that also have above-median mpg.

```python

cars_clean = cars.dropna(subset=['horsepower', 'weight', 'mpg', 'name'])

# Compute median mpg
median_mpg = cars_clean['mpg'].median()

# Filter and copy above-median mpg cars
above_median_mpg = cars_clean[cars_clean['mpg'] > median_mpg].copy()

# Calculate horsepower-to-weight ratio
above_median_mpg['hp_to_weight'] = above_median_mpg['horsepower'] / above_median_mpg['weight']

# Find max ratio
max_ratio = above_median_mpg['hp_to_weight'].max()

# Get car(s) with max ratio
best_cars = above_median_mpg[above_median_mpg['hp_to_weight'] == max_ratio]

# Show results
print(best_cars[['name', 'horsepower', 'weight', 'mpg', 'hp_to_weight']])
```

### Design a multi-line plot using Matplotlib or Seaborn that shows the evolution of average mpg over the years, separately for each origin

```python
import matplotlib.pyplot as plt

# Ensure no missing values in relevant columns
cars_clean = cars.dropna(subset=['mpg', 'model_year', 'origin'])

# Group by year and origin, then compute average mpg
avg_mpg = cars_clean.groupby(['model_year', 'origin'])['mpg'].mean().reset_index()

# Create multi-line plot
plt.figure(figsize=(10, 6))
sns.lineplot(data=avg_mpg, x='model_year', y='mpg', hue='origin', marker='o')

# Add titles and labels
plt.title('Average MPG Over Years by Origin')
plt.xlabel('Model Year')
plt.ylabel('Average MPG')
plt.legend(title='Origin')
plt.grid(True)
plt.tight_layout()
plt.show()
```

### Create a Seaborn scatterplot (or PairGrid) where:

X = horsepower

Y = weight

Color by: origin

Size by: mpg

Hue order = ['japan', 'europe', 'usa']

Add meaningful plot titles and axis titles.

```python
import matplotlib.pyplot as plt

# Ensure clean data
cars_clean = cars.dropna(subset=['horsepower', 'weight', 'mpg', 'origin'])


cars_clean = cars_clean.dropna(subset=['horsepower'])

# Create scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(
    data=cars_clean,
    x='horsepower',
    y='weight',
    hue='origin',
    size='mpg',
    palette='Set2',
    hue_order=['japan', 'europe', 'usa'],
    sizes=(20, 200),  # Adjust size scale
    alpha=0.7,
    edgecolor='black'
)

# Add titles and labels
plt.title('Horsepower vs Weight of Cars\n(Size = MPG, Color = Origin)', fontsize=14)
plt.xlabel('Horsepower')
plt.ylabel('Weight')
plt.legend(title='Origin', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.grid(True)
plt.show()
```

### We define a consistent car model as one that was produced over multiple years and had very low variation in mpg across those years (standard deviation < 1.0).

Tasks:

Identify car names that appear in more than one model_year.

For each such name, compute the standard deviation of mpg across years.

Return the car(s) with the lowest variation in mpg, among those with at least 2 appearances and std(mpg) < 1.0.

Report the model name(s), number of appearances, and the average mpg.

Bonus:
Sort the result by number of appearances (descending), then mpg (descending).

```python
# Ensure 'name' and 'mpg' are usable
cars_clean = cars.dropna(subset=['name', 'mpg', 'model_year'])

# Group by 'name' and check how many different years each appears in
name_year_counts = cars_clean.groupby('name')['model_year'].nunique()

# Filter to keep only cars that appear in more than one model year
multi_year_cars = name_year_counts[name_year_counts > 1].index

# Filter original DataFrame to only those cars
filtered_cars = cars_clean[cars_clean['name'].isin(multi_year_cars)]

# Group by car name and compute required stats
summary = filtered_cars.groupby('name').agg(
    mpg_std=('mpg', 'std'),
    appearances=('name', 'count'),
    avg_mpg=('mpg', 'mean')
).reset_index()

# Filter to only cars with low mpg variation
consistent_cars = summary[summary['mpg_std'] < 1.0]

# Bonus: sort by number of appearances (descending), then avg_mpg (descending)
consistent_cars_sorted = consistent_cars.sort_values(by=['appearances', 'avg_mpg'], ascending=[False, False])

# Final result
print(consistent_cars_sorted[['name', 'appearances', 'avg_mpg']])
```
